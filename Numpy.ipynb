## ***Numpy functions:***

**arange**

```python
#np.arange(start,end,steps) to create 1d array
ar_1d=np.arange(1,13)
print(ar_1d)
#---------------------------------
print("\n")
even_arr=np.arange(1,13,3)
print(even_arr)
```

**Linspace()**:- It defines the range and how many value i need btw the range

```python
# create a range and no of value btw the range

value=np.linspace(1,13,5)
print(value)
```

**reshape()**:-This function define the conversion of 1D array into mlti-dimensional array.

```python
# we have a 1D array
ar_1d=np.array([1,2,3,4,5,6])
print(ar_1d)
print("\n")

# now convert the 1D array into 2D array
ar_2d=ar_1d.reshape(2,3)
print("2D array is",ar_2d)
print("\n")

# now convert 1d array into 3D array and must focus on no elements in 1d arrya using which we will create multi array
ar_3d=ar_1d.reshape(2,3,1)
print("3D array is",ar_3d)
```

```python
#create 2d array from 1d array
ar=np.arange(1,13).reshape(2,6)
print(ar)

```

**ravel()**:-This function convert multi-dimensional array into 1d array

```python
#conversion of multi array into 1d arr using ravel fxn

arr_1d=ar.ravel()
print(arr_1d)
```

**flatten()**:-similarly this function is used to convert the multi array into 1d array

```python
#conversion of multi array into 1d arr using flatten fxn

print((ar.flatten()))
```

**transpose()**:-this fxn convert row into column

```python
mtrx=np.array([1,2,3,4,5,6])
print(mtrx.reshape(2,3))
print(mtrx.transpose())
```
